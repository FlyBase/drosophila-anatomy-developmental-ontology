```python
import pandas as pd
import csv
from collections import OrderedDict

#current type - which of the three sets?
current_set = "fascicles"

#load CSVs as pandas DataFrames
#these are the manual edits of autogenerated searches

#if current_set == "lineages":
    #l_mapping = pd.read_csv('spreadsheets/fb_lb_lineage_mapping.csv', sep='\t')# no terms yet
    #l_lbterms = pd.read_csv('spreadsheets/lb_lineages.tsv', sep='\t')# no terms yet
    #lbtype = "axon_tract"

if current_set == "compartments":
    mapping = pd.read_csv('spreadsheets/fb_lb_compartment_mapping.csv', sep='\t')
    lbterms = pd.read_csv('spreadsheets/lb_compartments.tsv', sep='\t')
    lbtype = "neuropil"

if current_set == "fascicles":
    mapping = pd.read_csv('spreadsheets/fb_lb_fascicle_mapping.csv', sep='\t')
    lbterms = pd.read_csv('spreadsheets/lb_fascicles.tsv', sep='\t')
    lbtype = "axon_tract"

mapping = mapping.applymap(str) # make everything in dataframe a string
lbterms = lbterms.applymap(str)
lbterms
```

```python
mapping = mapping.query('yes_no_approx != "N"')#filter out terms not matched to FBbt
```

```python
merged_table = pd.merge(left=lbterms, right=mapping, how='inner', on='larvalbrain_Name', sort=False)
merged_table = merged_table.applymap(str) # make all data string format
merged_table = merged_table.drop("notes", axis=1) #remove notes column
merged_table
```

```python
# Make a dictionary with key - column header & value = template specification (first row of table).
# Make first two columns

template_seed = OrderedDict([ ('ID' , 'ID'), ('CLASS_TYPE' , 'CLASS_TYPE'),\
                             ('RDF_Type' , 'TYPE' )])

#label, description, short synonym:

template_seed.update([("Xref" , "A oboInOwl:hasDbXref"), ("Name" , "A rdfs:label")])

# Create dataFrame for template
# from_records takes a list of dicts - one for each row.  We only have one row.

template = pd.DataFrame.from_records([template_seed])

template
```

```python
for i in merged_table.index:

    row_od = OrderedDict([]) #new template row as an empty ordered dictionary
    for c in template.columns: #make columns and blank data for new template row
        row_od.update([(c , "")])
    
    #these are the same in each row
    row_od["CLASS_TYPE"] = "subclass"
    row_od["RDF_Type"] = "owl:Class"
    
    #FBbt ID
    row_od["ID"] = merged_table.FBbt_id[i]
    
    #lookup larvalbrain ID and name
    #row_od["Xref"] = "http://www.larvalbrain.org/#/browse/" + lbtype + "/" + merged_table.larvalbrain_id[i]
    row_od["Xref"] = "larvalbrain_" + lbtype + ":" + merged_table.larvalbrain_id[i]
    row_od["Name"] = merged_table.FBbt_Name[i]
    
    #make new row into a DataFrame and add it to template
    new_row = pd.DataFrame.from_records([row_od])
    template = pd.concat([template, new_row], ignore_index=True, sort=False)

template
```

```python
template.to_csv("./template.tsv", sep = "\t", header=True, index=False)
```
